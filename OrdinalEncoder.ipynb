```python
import pandas as pd
import numpy as np

def ordinal_encode(data, ordinal_mapping):

    encoded_data = data.copy()

    for column, mapping in ordinal_mapping.items():

        encoded_data[column] = encoded_data[column].map(mapping)

    return encoded_data


data = pd.DataFrame({'Category': ['Low', 'Medium', 'High', 'Low', 'High', 'Medium'],
    'Status': ['Active', 'Inactive', 'Active', 'Inactive', 'Active', 'Inactive'], 'Rating': ['Good', 'Average', 'Good', 'Average', 'Bad', 'Bad']
})

ordinal_mapping = {'Category': {'Low': 1, 'Medium': 2, 'High': 3},
    'Status': {'Inactive': 0, 'Active': 1}, 'Rating': { 'Good': 2, 'Bad': 0, 'Average': 1 } }

encoded_data = ordinal_encode(data, ordinal_mapping)

print("Original Data:")
print(data)
print("\n\nEncoded Data:")
print(encoded_data)
```
